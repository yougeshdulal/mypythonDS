```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
a = np.array([3.5,3.5,1,2,4,6,7,4.5,4.5])
b = np.array([3.5,6.5,9,9,7,9,9,6.5,3.5])
plt.plot(a,b)
plt.show()
```

```python
a = np.array([1,5,11])
b = np.array([2,9,14])
plt.plot(a)
plt.plot(b)
plt.show()
```

```python
a = np.array([1,5,11])
b = np.array([2,9,14])
plt.plot(a,b,marker='X',ms = 5, mfc = 'r')
```

```python
plt.plot(a,b,marker='D',ms = 5)
```

```python
a = np.array([1,5,11])
b = np.array([2,9,14])

plt.title("This is test plot")
plt.xlabel("time taken")
plt.ylabel("Distance travelled")
plt.plot(a,b, marker='o',ms = 20, mfc = 'r',mec = 'k', linewidth = 3, linestyle = 'dotted')
plt.show()
```

```python
plt.bar(a,b, color = '000000')
plt.show()
```
